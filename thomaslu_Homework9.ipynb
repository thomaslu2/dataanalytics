**Problem 1.**  Its march madness!  Everyone likes to argue about which conference is best.  To shed some light, scrape data (using python and the Requests and BeautifulSoup libraries) from this website:
http://www.ncaa.com/rankings/basketball-men/d1/ncaa-mens-basketball-rpi

In Answer1, return a DataFrame of the extracted School, Conference, and Rank for all 351 schools, sorted by rank from best to worst.


```python
# Problem 1
# YOUR CODE HERE
#raise NotImplementedError()
url = 'http://www.ncaa.com/rankings/basketball-men/d1/ncaa-mens-basketball-rpi'
page = requests.get(url)
soup = bs4.BeautifulSoup(page.content,'html.parser')
table = soup.find('table',{'class':'sticky'})
rows = table.findAll('tr')
data = []
tablesch, tableconf, tablerank = ([] for i in range(3))
for tr in rows:
    data.append(tr.findAll('td'))
for i in range(1,352):
    temp = data[i][2].text
    temp = temp.strip(";")
    if(temp == 'A&M-Corpus; Christi'):
        temp = 'A&M-Corpus Christi'
    tablesch.append(temp)
    tableconf.append(data[i][3].text)
    tablerank.append(int(data[i][0].text))
Answer1 = pandas.DataFrame({
    "School": tablesch,
    "Conference": tableconf,
    "Rank": tablerank
})
Answer1
```

```python
# Problem 1 autograder -- run me
nose.tools.assert_is_instance(Answer1, pandas.DataFrame)
nose.tools.assert_tuple_equal(Answer1.shape, (351, 3))
nose.tools.assert_list_equal(list(Answer1.columns), ['School','Conference','Rank'])
nose.tools.assert_true(Answer1.Rank.is_monotonic_increasing)
# autograder hidden tests
'passed'
```

**Problem 3.** 
Web-scraping is especially useful for integrating data together from many web pages.  This page has links to each team's stats in 2018:
http://www.sports-reference.com/cbb/seasons/2018-school-stats.html

Scrape the team names and links.

In Answer3, return a list of tuples containing each School name and its URL, in the same list order as on the web page.  Use the relative URLs as given in the page.

```python
# Problem 3

# YOUR CODE HERE
#raise NotImplementedError()
url = 'https://www.sports-reference.com/cbb/seasons/2018-school-stats.html'
page = requests.get(url)
soup = bs4.BeautifulSoup(page.content,'html.parser')
table = soup.find('table',{'class':'sortable'})
data = table.findAll('a')
links = []
schools3 = []
tempanswer = []
for i in data:
    if i.has_attr('href'):
        links.append(i.attrs['href'])
        schools3.append(i.text)
for j in range(len(schools3)):
    tempanswer.append((schools3[j],links[j]))
Answer3 = tempanswer
tuple(Answer3)
```

```python
# Problem 3 autograder -- run me
nose.tools.assert_is_instance(Answer3, list)
nose.tools.assert_tuple_equal(Answer3[0], ('Abilene Christian', '/cbb/schools/abilene-christian/2018.html'))
# autograder hidden tests
'passed'
```
