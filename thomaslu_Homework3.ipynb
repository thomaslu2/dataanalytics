```python
# Problem 1
# YOUR CODE HERE
survey.Major.replace({'CS':'Computer Science', 'CMDA':'Computational Modeling and Data Analytics', 'CpE':'Computer Engineering'},inplace=True)
import numpy as np
survey.YearsVT.replace({"Six":6,"3 years":3, "3.5":3,"2 (starting my 3rd)":2,"4 (on 5th currently)":4,"3rd year at Tech":3},inplace=True)
survey.Name.replace({"bethel":"Bethel", "Thomas Lu":"Thomas","Pavani Surapaneni":"Pavani","JINGYI":"Jingyi","Deniz Atak":"Deniz","Fuadul Islam":"Fuadul"},inplace=True)
survey.Name.replace({"Junyuan":"Junyuan","Tiffany Ma":"Tiffany", "William (Bob)":"Bob", "Armani McKinnie":"Armani","Youssef Salah":"Youssef","John Doe":"John"},inplace=True)
#survey.Gender.replace(np.nan,"Male", inplace=True)
survey.Height.replace({"5'2":62,"5'5":65,"5'8\"":68,"5'9":69,"5' 10":70,"5'6":66,"6'0":72,"5'11":71,"5'6\"":66},inplace=True)
survey.DistanceBorn = survey.DistanceBorn.map(lambda v: int(v.replace(",","").replace("miles","")))
survey.BedTime.replace(np.nan,12,inplace=True)
survey.BedTime.replace({"12.5":12,"11:00":11,"10.5":10,"23.5":11,"12:30 AM":12,"2AM":14,"11:30":11,"13.5":13,"1:00":13},inplace=True)
mean = int(survey.PhoneMins.mean())
survey.PhoneMins.replace(np.nan,mean,inplace=True)
survey.PhoneApps.replace(np.nan,1670,inplace=True)
survey.PhoneApps.replace({'20+':20},inplace=True)
survey.Photos.replace({'2,000':2000, '500+':500},inplace=True)
survey.Photos.replace(np.nan,724,inplace=True)
survey.Facebook.replace({'1,477':1477},inplace=True)
survey.Facebook.replace(np.nan,447,inplace=True)
survey.Temperature.replace('60 to 70 range',65,inplace=True)
survey.Spicy.replace('87.5',87,inplace=True)
survey.Football.replace('dont remember', 8,inplace=True)
survey.HardDrive.replace({'110 GB':1100000000000,'8':np.NaN,'167':np.nan,'435300':np.NaN,'160 GB':1600000000000,'64':np.NaN,'67':np.NaN,'400GB':4000000000000,'190GB':1900000000000,'120':np.NaN,'233 GB':2330000000000,'180 GB':1800000000000,'313GB':3130000000000,'0':np.NaN,'200':np.NaN,'6':np.NaN,'64':np.NaN,'5000000':np.NaN,'120GB':1200000000000,'200 GB':2000000000000,'172 GB':1720000000000,'10 GB':100000000000,'5':np.NaN},inplace=True)
#survey.HardDrive.replace({'190GB':190000000000,'120':np.NaN,'233 GB':233000000000,'180 GB':180000000000,'313GB':313000000000,'0':np.NaN},inplace=True)
#survey.HardDrive.replace({'200':np.NaN,'6':np.NaN,'64':np.NaN,'5000000':np.NaN,'120GB':120000000000,'200 GB':200000000000,'172 GB':172000000000,'10 GB':10000000000,'5':np.NaN},inplace=True)
#survey.HardDrive.fillna(0,inplace=True)
survey.HardDrive = survey.HardDrive.astype(np.float64)
survey.YearsVT = survey.YearsVT.astype(np.int64)
survey.Height = survey.Height.astype(np.int64)
survey.DistanceBorn = survey.DistanceBorn.astype(np.int64)
survey.BedTime = survey.BedTime.astype(np.int64)
survey.PhoneMins = survey.PhoneMins.astype(np.int64)
survey.PhoneApps = survey.PhoneApps.astype(np.int64)
survey.Photos = survey.Photos.astype(np.int64)
survey.Facebook = survey.Facebook.astype(np.int64)
survey.Temperature = survey.Temperature.astype(np.int64)
survey.Spicy = survey.Spicy.astype(np.int64)
survey.Football = survey.Football.astype(np.int64)
#raise NotImplementedError()

# Assign your result to variable Answer1.

Answer1 = survey
```

**Problem 2:**  What is the average Height of Males and Females in the class?  Using the cleaned data, return the result in a groupby table as Answer2.

```python
# Problem 2
# YOUR CODE HERE
#raise NotImplementedError()

# Assign your result to variable Answer2.

Answer2 = survey.groupby(['Gender'])['Height'].mean()
```

**Problem 3:**  What is the Major x Gender population distribution in our class?
Display the frequency counts of each Major x Gender group in a pivot table format (a table with columns for Genders and rows for Majors). Sort the Majors and Genders alphabetically. Hint:  multi-level grouping, unstack, pivot_table might be helpful.
Return the pivot table as Answer3.

```python
# Problem 3
# YOUR CODE HERE
#raise NotImplementedError()
# Assign your result to variable Answer3.

Answer3 = survey.groupby(['Major','Gender'])['Major'].count().unstack()
print(Answer3)
print(Answer3.shape)
```
