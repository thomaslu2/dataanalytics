**Problem 1:** Who are the biggest outliers in each column?  
Model an outlier metric as the absolute value of the z-score (i.e. normalized distance from the column mean). For each column, find the person who is the most outlier in that column, and what their original (survey) value for that column is. Hint: you might find idxmax() helpful.

In Answer1, return a List of tuples of the form: \[(Column name, Person name, Survey value), ...\], sorted alphabetically by Column name.

```python
# Problem 1
# YOUR CODE HERE
#raise NotImplementedError()
tempvar = Answer0.copy()
temp1 = []
temp2 = []
for column in tempvar.columns:
    tempvar[column] = (abs(tempvar[column]) - tempvar[column].mean())
    temp1.append(tempvar[column].idxmax())
for i in range(len(temp1)):
    temp2.append((tempvar.columns.values[i], table.Name[temp1[i]], table[tempvar.columns[i]][temp1[i]]))
Answer1 = temp2
```

**Problem 3:** What is the multi-dimensional outlier-ness of each person?
In data analytics, we often try to quantitatively model qualitative concepts that involve multiple dimensions. For example, we may have an intuitive idea of the total "outlier-ness" of a person. Now we want to model the total multi-dimensional outlierness of each person as a single metric that takes into account all the survey questions.

Model the total outlierness of each person as the root-mean-square of that person's whole row of z-scores (hint: the axis parameter on pandas reduce operators is handy sometimes):
![formula](https://wikimedia.org/api/rest_v1/media/math/render/svg/fc621ce0b9b2d52e3ce835a9211f042f272c341e)

In Answer3, return a DataFrame containing two columns ('Name', 'Outlierness'), sorted by Outlierness from most to least.

Think: Did you land where you expected in the list?  
Note: You just computed Euclidean distance of each person to the center of mass, also known as Mahalanobis distance https://en.wikipedia.org/wiki/Mahalanobis_distance .

```python
# Problem 3
# YOUR CODE HERE
#raise NotImplementedError()
rms = Answer0.applymap(numpy.square)
templist = []
templist = (rms.sum(axis=1)/30) ** 0.5
new_dataframe = pandas.DataFrame(
    {
       "Name": table.Name,
        "Outlierness": templist
    }
)
Answer3 = new_dataframe.sort_values('Outlierness',ascending=False)

```

**Problem 5:** What are the most positively and negatively correlated columns?

In Answer5, return a list of column pairs, sorted by the pearson correlation values from most positive to most negative, in the form:  \[(Column name1, Column name2, Correlation value), ...].  Do not include any pairs twice, do not include column self correlations, and name1 should alphabetically precede name2.

```python
# Problem 4
# YOUR CODE HERE
#raise NotImplementedError()
Answer5 = []
goodwin = Answer0.corr().stack().reset_index()
goodwin.columns = ['Coll','Col2','Cor']
for i in range(len(goodwin.index)):
    if(goodwin['Coll'][i] != goodwin['Col2'][i]):
        Answer5.append((goodwin['Coll'][i], goodwin['Col2'][i], goodwin['Cor'][i]))
for i in reversed(Answer5):
    freq = 0
    for j in reversed(Answer5):
        if(j[2] == i[2]):
            freq = freq + 1
    if (freq > 1):
        Answer5.remove(i)
Answer5.sort(key=lambda tup: tup[2], reverse=True)
Answer5
```
